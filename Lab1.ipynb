```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import warnings

warnings.filterwarnings("ignore")
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([52, 55, 61, 70, 82])
```

#Train the model_A

```python
model_1 = LinearRegression()
model_1.fit(X, y)
print("Slope:", model_1.coef_)
print("Intercept:", model_1.intercept_)
```

```python
print("Model_1 Predictions:", model_1.predict(X))
```

```python
x_test = np.array([[6]])
y_pred_1 = model_1.predict(x_test)

print("Model A Prediction at x=6:", y_pred_1)
```

```python
y_train_pred_1 = model_1.predict(X)
mse_1 = mean_squared_error(y, y_train_pred_1)

print("MSE of Model A:", mse_1)
```

#Train the model_B

```python
x_poly = X.flatten()
coeff = np.polyfit(x_poly, y, 4)
print("Polynomial coefficients:", coeff)
```

```python
poly_model = np.poly1d(coeff)
```

```python
y_pred_B = poly_model(6)
print("Model B Prediction at x=6:", y_pred_B)
```

```python
y_tain_pred_2 = poly_model(X.flatten())
mse_2 = mean_squared_error(y, y_tain_pred_2)
print("MSE of Model B:", mse_2)
```

#Model_1 has higher bias and Model_2 has Higher variance
